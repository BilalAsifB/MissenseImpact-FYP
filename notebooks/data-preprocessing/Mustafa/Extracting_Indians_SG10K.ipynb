```python
# ==============================================
# Purpose: Extract all aliases for samples with region = "Indian"
# ==============================================

# -------------------------------
# 1. Import required libraries
# -------------------------------
import pandas as pd
import json

# -------------------------------
# 2. Load CSV data into DataFrame
# -------------------------------
df = pd.read_csv("samples.csv")  # CSV containing sample metadata

# -------------------------------
# 3. Extract 'region' from JSON in 'extra_attributes'
# -------------------------------
def get_region(attr):
    """
    Parse the 'extra_attributes' JSON and return the value for 'region' tag.

    Parameters:
        attr (str): JSON string from 'extra_attributes' column

    Returns:
        str or None: value of region, or None if not found or JSON invalid
    """
    try:
        for item in json.loads(attr):
            if item.get("tag") == "region":
                return item.get("value")
    except Exception:
        return None

# Apply the function to create a new 'region' column
df["region"] = df["extra_attributes"].apply(get_region)

# -------------------------------
# 4. Inspect data (optional)
# -------------------------------
# Count how many samples have region == "Indian"
indian_count = (df['region'] == 'Indian').sum()
print(f"Number of samples with region='Indian': {indian_count}")

# See all unique regions present
print("Unique regions in dataset:", df['region'].unique())

# -------------------------------
# 5. Filter DataFrame for region = "Indian"
# -------------------------------
df_indian = df.loc[df["region"] == "Indian", ["alias", "region"]]

# -------------------------------
# 6. Keep only the 'alias' column
# -------------------------------
df_indian_alias = df_indian["alias"]

# -------------------------------
# 7. Save aliases to plain text file
# -------------------------------
# Each alias on a new line, no header or index
df_indian_alias.to_csv("indian_alias.txt", index=False, header=False)

print("Aliases for region='Indian' saved to df_indian_alias.txt")
```
