```python
import os
import subprocess

print('Imports ready.')
```

```python
BASE = '/content/drive/MyDrive/FYP_DATA'
ANNOTATED_MANE_DIR = f'{BASE}/DATA/annotated_data_mane'

# Chromosome list (1-22)
CHROMOSOMES = [str(i) for i in range(1, 23)]

# Dataset path templates
DATASET_PATHS = {
    'sg10k': f'{ANNOTATED_MANE_DIR}/sg10k/sg10k_annotated/',
    'gnomad': f'{ANNOTATED_MANE_DIR}/gnomad/gnomad_pure_sas_annotated/',
    '1000genomes': f'{ANNOTATED_MANE_DIR}/1k/1k_annotated/',
}

print('Configuration loaded.')
print(f'CHROMOSOMES: {CHROMOSOMES}')
```

```python
print("⏳ Indexing all annotated sg10k mane annotated files...")
print("=" * 60)

results = {}

for chr_num in range(1,23):

  vcf_file = os.path.join(DATASET_PATHS['sg10k'], f'sg10k_chr{chr_num}_SAS_annotated_mane.vcf.gz')
  tbi_file = vcf_file + '.tbi'

  if os.path.exists(tbi_file):
    print(f"⏭️  chr{chr_num:2d}: Already indexed — skipping")
    results[chr_num] = 'skipped'
    continue

  # Check input exists
  if not os.path.exists(vcf_file):
    print(f"❌ chr{chr_num:2d}: File not found — {vcf_file}")
    results[chr_num] = 'missing'
    continue

  # Index
  print(f"🔄 chr{chr_num:2d}: Indexing...")
  result = subprocess.run(
      f"bcftools index -t {vcf_file}",
      shell=True, capture_output=True, text=True
      )

  if result.returncode == 0:
    tbi_size = os.path.getsize(tbi_file) / (1024**2)
    print(f"✅ chr{chr_num:2d}: Done ({tbi_size:.2f} MB index)")
    results[chr_num] = 'done'
  else:
    print(f"❌ chr{chr_num:2d}: Failed — {result.stderr.strip()[:100]}")
    results[chr_num] = 'failed'

# ── Summary ──────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("INDEXING SUMMARY")
print("=" * 60)
print(f"✅ Done    : {sum(1 for v in results.values() if v == 'done')}")
print(f"⏭️  Skipped : {sum(1 for v in results.values() if v == 'skipped')}")
print(f"❌ Failed  : {sum(1 for v in results.values() if v == 'failed')}")
print(f"❓ Missing : {sum(1 for v in results.values() if v == 'missing')}")

missing = [k for k, v in results.items() if v == 'missing']
failed  = [k for k, v in results.items() if v == 'failed']
if missing:
    print(f"\n   Missing chromosomes : {missing}")
if failed:
    print(f"   Failed chromosomes  : {failed}")
print("=" * 60)
```
