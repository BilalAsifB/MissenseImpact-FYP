# Rename all files, for readability

```python
import os

BASE_DIR = '/content/drive/My Drive/FYP_DATA/DATA/annotated_data/sg10k/annotated/sg10k_annotated'

renamed = 0
missing = 0
skipped = 0

for i in range(1, 23):
    old_path = os.path.join(BASE_DIR, f"sg10k_chr{i}_SAS_annotated.vcf.gz")
    new_path = os.path.join(BASE_DIR, f"sg10k_chr{i}_annotated.vcf.gz")

    if not os.path.exists(old_path):
        print(f"⚠️  Missing: chr{i}")
        missing += 1
        continue

    if os.path.exists(new_path):
        print(f"⏭️  Already exists, skipping: chr{i}")
        skipped += 1
        continue

    os.rename(old_path, new_path)
    print(f"✅ Renamed: chr{i}")
    renamed += 1

print("\n" + "="*60)
print(f"Renamed : {renamed}")
print(f"Skipped : {skipped}")
print(f"Missing : {missing}")
print("="*60)
```

# Indexing all vcf's

```python
import subprocess
import os

ANNOTATED_DIR = '/content/drive/MyDrive/FYP_DATA/DATA/annotated_data_mane/indigen/indigen_annotated/'

print("⏳ Indexing all annotated sg10k files...")
print("=" * 60)

results = {}

for chr_num in range(1, 23):
    vcf_file  = os.path.join(ANNOTATED_DIR, f'indigen_annotated_mane.vcf{chr_num}_annotated.vcf.gz')
    tbi_file  = vcf_file + '.tbi'

    # Skip if already indexed
    if os.path.exists(tbi_file):
        print(f"⏭️  chr{chr_num:2d}: Already indexed — skipping")
        results[chr_num] = 'skipped'
        continue

    # Check input exists
    if not os.path.exists(vcf_file):
        print(f"❌ chr{chr_num:2d}: File not found — {vcf_file}")
        results[chr_num] = 'missing'
        continue

    # Index
    print(f"🔄 chr{chr_num:2d}: Indexing...")
    result = subprocess.run(
        f"bcftools index -t {vcf_file}",
        shell=True, capture_output=True, text=True
    )

    if result.returncode == 0:
        tbi_size = os.path.getsize(tbi_file) / (1024**2)
        print(f"✅ chr{chr_num:2d}: Done ({tbi_size:.2f} MB index)")
        results[chr_num] = 'done'
    else:
        print(f"❌ chr{chr_num:2d}: Failed — {result.stderr.strip()[:100]}")
        results[chr_num] = 'failed'

# ── Summary ──────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("INDEXING SUMMARY")
print("=" * 60)
print(f"✅ Done    : {sum(1 for v in results.values() if v == 'done')}")
print(f"⏭️  Skipped : {sum(1 for v in results.values() if v == 'skipped')}")
print(f"❌ Failed  : {sum(1 for v in results.values() if v == 'failed')}")
print(f"❓ Missing : {sum(1 for v in results.values() if v == 'missing')}")

missing = [k for k, v in results.items() if v == 'missing']
failed  = [k for k, v in results.items() if v == 'failed']
if missing:
    print(f"\n   Missing chromosomes : {missing}")
if failed:
    print(f"   Failed chromosomes  : {failed}")
print("=" * 60)
```
